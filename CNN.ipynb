```python
import numpy as np
import pandas as pd
import seaborn as sbn
import matplotlib.pyplot as plt
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Conv2D,MaxPooling2D,Activation,Dense,Dropout,Flatten
from keras.metrics import mean_squared_error
train_path = "../input/fashionmnist/fashion-mnist_train.csv"
test_path = "../input/fashionmnist/fashion-mnist_test.csv"
train_df = pd.read_csv(train_path)
test_df = pd.read_csv(test_path)
```

```python
train_value=train_df.iloc[:,1:].values
test_value=test_df.iloc[:,1:].values
```

```python
train_value=train_value.reshape(len(train_value),28,28,1)
test_value=test_value.reshape(len(test_value),28,28,1)
```

```python
for i in np.arange(6):
    plt.figure()
    plt.imshow(train_value[i]);
    plt.axis("off");
```

```python
for i in np.arange(5):
    plt.figure()
    plt.imshow(test_value[i]);
    plt.axis("off");
```

```python
def lap(data_path):
    data = pd.read_csv(data_path)
    data = data.values
    np.random.shuffle(data)
    x = data[:,1:].reshape(-1,28,28,1)/255.0
    y = data[:,0].astype(np.int32)
    y = tf.keras.utils.to_categorical(y,num_classes=len(set(y)))
    
    return x,y

x_train,y_train = lap("../input/fashionmnist/fashion-mnist_train.csv")
x_test,y_test = lap("../input/fashionmnist/fashion-mnist_test.csv")
```

```python
## Creating CNN
model = Sequential()
model.add(Conv2D(12,(1,1)))
model.add(Activation("relu"))
model.add(MaxPooling2D())

model.add(Conv2D(12,(1,1)))
model.add(Activation("relu"))
model.add(MaxPooling2D())

model.add(Conv2D(12,(1,1)))
model.add(Activation("relu"))
model.add(MaxPooling2D())

## entering NN with important features from CNN

model.add(Flatten())
model.add(Dense(10)) 
model.add(Activation("relu"))
model.add(Dense(10))
model.add(Activation("relu"))
model.add(Dense(10))
model.add(Activation("relu"))
model.add(Dense(10))
model.add(Activation("relu"))
model.add(Dense(10))
model.add(Activation("relu"))
model.add(Dense(10))
model.add(Activation("softmax"))
## model compiler
model.compile(loss="categorical_crossentropy",
             optimizer="rmsprop",
             metrics=["accuracy"])
hist=model.fit(x=x_train,y=y_train,validation_data=(x_test,y_test),epochs=50,batch_size=600)
```

```python
model.summary()
```

```python
for i in hist.history.keys():
    plt.figure()
    plt.plot(hist.history[i]);
    plt.title(str(i))
```
